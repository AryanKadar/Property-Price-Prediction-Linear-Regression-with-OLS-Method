```python
y=t1.iloc[:,-1]
y
```

```python
xtrain,xtest,ytrain,ytest=TTS(x,y,test_size=0.2,random_state=2)
```

```python
mkp=MKP(SS(),LR())
```

```python
mkp.fit(xtest,ytest)
```

```python
mkp.score(xtrain,ytrain)
```

```python
CVS(mkp,xtrain,ytrain,cv=6,scoring="r2")
```
